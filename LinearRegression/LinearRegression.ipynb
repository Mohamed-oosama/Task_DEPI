```python
import numpy as np


class LinearRegression:

    def __init__(self, lr=0.01, epochs=1000):    # default the number to start
        self.lr = lr
        self.epochs = epochs
        self.w = None
        self.b = None
        self.loss_history = []

    def fit(self, x, y):   # translate the data to array
        x = np.array(x)
        y = np.array(y)

        n_samples, n_features = x.shape     # know the number the array

        self.w = np.zeros(n_features)
        self.b = 0

        for epoch in range(self.epochs):

            y_pred = np.dot(x, self.w) + self.b     # cala the number the y pred

            # gradients
            dw = (2 / n_samples) * np.dot(x.T, (y_pred - y))   # update the number weight
            db = (2 / n_samples) * np.sum(y_pred - y)          # update the number bias

            # update
            self.w -= self.lr * dw
            self.b -= self.lr * db

            # MSE
            loss = self.mse(y, y_pred)
            self.loss_history.append(loss)


    def predict(self, x):
        x = np.array(x)

        if len(x.shape) == 1:
            x = x.reshape(-1, 1)

        return np.dot(x, self.w) + self.b

    # Mean Squared Error
    def mse(self, y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    # Root Mean Squared Error
    def rmse(self, y_true, y_pred):
        return np.sqrt(self.mse(y_true, y_pred))

    # R² Score
    def r2_score(self, x, y):
        y = np.array(y)
        y_pred = self.predict(x)

        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)

        return 1 - (ss_res / ss_tot)
```
